```python
from bs4 import BeautifulSoup
import requests
import csv
url = "https://www.worldometers.info/coronavirus/"
html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")
```

```python
soup.title.string
```

```python
tr = soup.find_all('tr')
```

```python
def extract_text(row, tag):
    element = BeautifulSoup(row, 'html.parser').find_all(tag)
    text = [col.get_text() for col in element]
    return text
```

```python
heading = tr.pop(0)
```

```python
heading_row = extract_text(str(heading), 'th')
```

```python
heading_row = heading_row[1:9]
```

```python
print(heading_row)
```

```python
len(heading_row)
```

```python
tds = [extract_text(str(row), 'td')[1:9] for row in tr]
```

```python
import pandas as pd
```

```python

df = pd.DataFrame(tds)
```

```python
df.to_csv('corona.csv')
```
